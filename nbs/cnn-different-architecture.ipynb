```python
#Loading the data
DATA = untar_data(URLs.PETS)
path = Path('/root/.fastai/data/oxford-iiit-pet/images')
```

```python
#Simple function to get correct labels
#Pets stored as name_number.jpg
#removing number and .jpg
def pet_label_func(fn):
    x_split = fn.split("_")
    if len(x_split) == 2:
        return x_split[0]
    else:
        label = ""
        for i in range(len(x_split) - 2):
            label += x_split[i] + " "
        label += x_split[-2]
        return label
```

```python
dls = ImageDataLoaders.from_name_func(
    path, get_image_files(path), valid_pct=0.2,
    label_func=pet_label_func, item_tfms=[Resize(80, method='crop')])
dls.show_batch(max_n=6)
```

```python
def get_plot_data(dls, modelname): 
    model = vision_learner(dls, modelname, metrics=error_rate, cbs=CSVLogger())
    model.fine_tune(10)
    df = model.csv_logger.read_log() 
    return df
```

```python
resnet18 = get_plot_data(dls,'resnet18')
```

```python
resnet26 = get_plot_data(dls,'resnet26')
```

```python
resnet34 = get_plot_data(dls,'resnet34')
```

```python
efficientnet_b1 = get_plot_data(dls,'efficientnet_b1')
```

```python
resnet26t = get_plot_data(dls,'resnet26t')
```

```python
resnetblur50 = get_plot_data(dls,'resnetblur50')
```

```python
csvs = [resnet18, resnet26,resnet34, efficientnet_b1, resnet26t, resnetblur50]
plt.figure(figsize = (8,8))
for i in csvs:
    plt.plot(i['epoch'], i['error_rate'])
    plt.xticks(i['epoch'])

plt.xlabel("Epoch")

plt.ylabel("Error rate")
plt.legend(['resnet18','resnet26','resnet34','efficientnet_b1', 'resnet26t', 'resnetblur50'])

plt.title("CNN with different architecture")
```
